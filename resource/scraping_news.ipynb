```python
df_list=[]
base_url=('http://newsapi.org/v2/top-headlines?'
     'language=en&'
     'apiKey=f89db8919ad3464bae1e8a9b39c19164')
print(f'Fetchihng {base_url}...')
response=requests.get(base_url)
data=response.json()
df_list.append(pd.DataFrame(data['articles']))
status=data['status']
total_counts=min(data['totalResults'], 100) # // basic account limited to 100 max results
pages=-(-total_counts//20) # // rounds down
for each_page in range(2, pages+1): # // page 1 already scraped
    url=base_url+f'&page={each_page}'
    print(f'Fetchihng {url}...')
    response=requests.get(url)
    data=response.json()
    df_list.append(pd.DataFrame(data['articles']))
#     break
# combined_df=pd.concat(df_list, ignore_index='True')
# combined_df
```

```python
df=pd.concat(df_list, ignore_index='True')
df['source']=df['source'].apply(lambda x: x['name'])
```
