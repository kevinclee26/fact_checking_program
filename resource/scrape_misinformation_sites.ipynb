## grab all extensions from [Wiki](https://en.wikipedia.org/wiki/List_of_Internet_top-level_domains#Original_top-level_domains). 

```python
import requests
from bs4 import BeautifulSoup
import pandas
import csv
```

```python
tables=pd.read_html('https://en.wikipedia.org/wiki/List_of_Internet_top-level_domains#Original_top-level_domains')
domain_list=[]
for each_table in tables: 
    if 'Name' in list(each_table.columns): 
        for each_domain in each_table['Name'].iteritems(): 
            domain_list.append(each_domain[1])
domain_list
```

## grab misinformation directory from factcheck.org

```python
url='https://www.factcheck.org/2017/07/websites-post-fake-satirical-stories/'
response=requests.get(url)
soup=BeautifulSoup(response.text, 'html.parser')
all_text=soup.find('div', class_='entry-content').text
soup=BeautifulSoup(response.text, 'html.parser')
all_text=soup.find('div', class_='entry-content').text
all_tokens=all_text.replace('\n', ' ').replace('“', ' ').split(' ')
misinformation_sites=[]
for each_token in all_tokens: 
    if each_token.lower().endswith(tuple(domain_list)): 
        misinformation_sites.append(each_token)
misinformation_sites
```

```python
with open('data/misinformation_sites.csv', 'w') as f: 
    csv_writer=csv.writer(f, delimiter='\n') 
    csv_writer.writerow(misinformation_sites)
```
