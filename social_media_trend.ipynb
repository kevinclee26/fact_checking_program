```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('resource/data/users-by-social-media-platform.csv')
df.head()
```

```python
df[['Entity', 'Year', 'Monthly active users (Statista and TNW (2019)) (MAU)']].pivot(index='Year', columns='Entity', values='Monthly active users (Statista and TNW (2019)) (MAU)').plot().legend(bbox_to_anchor=(1, 1))
```

```python
graph_df=df[['Entity', 'Year', 'Monthly active users (Statista and TNW (2019)) (MAU)']]
graph_df['Monthly Active Users (Billions)']=graph_df['Monthly active users (Statista and TNW (2019)) (MAU)']/1000000000
graph_df[['Entity', 'Year', 'Monthly Active Users (Billions)']].pivot(index='Year', columns='Entity', values='Monthly Active Users (Billions)').interpolate().plot().legend(bbox_to_anchor=(1, 1))
plt.title('Number of People Using Social Media Platforms')
plt.ylabel('Monthly Active Users (Billions)')
plt.xlabel('Year')
```
